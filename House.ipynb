```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = datasets.fetch_california_housing()
```

```python
house
```

```python
df = pd.DataFrame(house.data, columns = house.feature_names)
```

```python
df.head()
```

```python
df['MedHouseVal'] = house.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
correlation = df.corr()
sns.heatmap(correlation, annot=True, square=True, fmt ='.1f', cmap='Greens', cbar=True)
```

```python
X = df.drop(columns='MedHouseVal',axis=1)
Y = df.MedHouseVal
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, Y.shape, X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
pred = model.predict(X_train)
```

```python
print(r2_score(pred,Y_train))
```

```python
model.fit(X_test, Y_test)
```

```python
pred = model.predict(X_test)
```

```python
print(r2_score(pred,Y_test))
```
